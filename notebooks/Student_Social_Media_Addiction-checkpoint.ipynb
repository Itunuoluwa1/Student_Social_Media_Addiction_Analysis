```python
# Average daily usage by gender
usage_by_gender = df.groupby('Gender')['Avg_Daily_Usage_Hours'].mean().sort_values()
sns.barplot(x=usage_by_gender.index, y=usage_by_gender.values)
plt.title('Average Daily Usage by Gender')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Average daily usage by age
usage_by_age = df.groupby('Age')['Avg_Daily_Usage_Hours'].mean().sort_values()
sns.barplot(x=usage_by_age.index, y=usage_by_age.values)
plt.title('Average Daily Usage by Age')
plt.show()
```

```python
# Average daily usage by academic level Pivot Table
academic_level_by_usage = df.pivot_table(
    values = 'Avg_Daily_Usage_Hours',
    index = 'Academic_Level',
    aggfunc = 'mean'
)
print(academic_level_by_usage)
```

```python
# Average daily usage by academic level Plot
academic_level_by_usage.plot(kind = 'bar', figsize = (7, 4))
plt.title('Average Daily Usage By Academic Level')
plt.show()
```

```python
# Sleep vs Addiction
sns.scatterplot(data=df, x='Addicted_Score', y='Sleep_Hours_Per_Night', hue='Gender')
plt.title('Addiction vs Sleep Hours')
plt.show()
```
